## Classification Model Evaluation

### import packages

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('Social_Network_Ads.csv')
print(df.head())
```

### EDA

```python
print(df.describe())
```

```python
print(df.info())
```

```python
print(df.isna().sum())
```

```python
# custom logic to convert textual to numerical
unique_gender_values = df['Gender'].unique()
replacement_values = [1, 2]
df = df.replace(unique_gender_values, replacement_values)
print(df.head())
```

```python
sns.pairplot(df)
```

### Model

```python
# decide x and y

x = df.drop(['Purchased', 'User ID', 'Gender'], axis=1)
y = df['Purchased']
```

```python
# split the dataset into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=123456)
```

```python
from sklearn.linear_model import LogisticRegressionCV

# create the model
model = LogisticRegressionCV()

# train the model
model.fit(x_train, y_train)
```

### Model Evaluation

```python
# predict the values
y_prediction = model.predict(x_test)
```

#### confusion matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_prediction)
print(cm)
```

#### accuracy

```python
accuracy = (cm[0, 0] + cm [1, 1]) / (cm[0, 0] + cm[0, 1] + cm [1, 1] + cm[1, 0])
print(f"accuracy = {accuracy * 100: 0.2f}%")
```

```python
from sklearn.metrics import accuracy_score

print(f"accuracy = {accuracy_score(y_test, y_prediction) * 100: 0.2f}%")
```

#### precison score

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_prediction)
print(f"precision = {precision}")
```

#### recall score

```python
from sklearn.metrics import recall_score

recall = recall_score(y_test, y_prediction)
print(f"recall = {recall}")
```

#### f1 score

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_prediction)
print(f"f1 score = {f1}")
```

#### classification report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_prediction))
```
